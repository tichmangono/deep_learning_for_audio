```python
import librosa, librosa.display
import os
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
import json
```

```python
#list(os.walk("data_gtzan/"))
```

```python
DATASET_PATH = "data_gtzan/genres_original/"
JSON_PATH = "data.json"
SAMPLE_RATE = 22050
DURATION = 30
SAMPLES_PER_TRACK = SAMPLE_RATE*DURATION
```

```python
def save_mfcc(dataset_path, json_path, n_mfcc=13, n_fft = 2048,
              hop_length=512, num_segments=5):
    # dict to store data
    data= {
        "mapping": [],
        "mfcc": [],
        "labels": []
    }
    
    num_samples_per_segment = int(SAMPLES_PER_TRACK/num_segments)
    expected_num_mfcc_vectors_per_segment = np.ceil(num_samples_per_segment/hop_length)
    
    # loop through all genres
    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):
        
        # ensure that we are not at root level
        if dirpath is not dataset_path:
            
            # save the semantic label
            dirpath_components = dirpath.split("/") # genre/blues=> ["genre", "blues"]
            semantic_label = dirpath_components[-1]
            data["mapping"].append(semantic_label)
            print(f"\n-----\nProcessing => {semantic_label}\n-----\n")
            # process files for specifc genre
            for f in filenames:
                
                #load audio file
                file_path = os.path.join(dirpath, f)
                try:
                    signal, sr = librosa.load(file_path, sr = SAMPLE_RATE)
                except Exception as e:
                    print(f"Failed to load {file_path} with {e}")
                    with open("errors.txt", "a") as err_file:
                        err_file.write(f"Failed to load {file_path} with {e}")
                    
                print(f"loaded audio filepath {file_path} file {f} length {len(signal)}")
                
                # process segments exctracting mfcc and store data
                for s in range(num_segments):
                    start_sample = num_samples_per_segment * s # s=0 -> 0
                    end_sample = start_sample + num_samples_per_segment # s=0 -> num_samples_per_segment
                        
                    mfcc = librosa.feature.mfcc(signal[start_sample:end_sample],
                                               sr = sr,
                                               n_fft = n_fft,
                                               n_mfcc=n_mfcc,
                                               hop_length= hop_length)
                    mfcc = mfcc.T
                    
                    # may have slightly longer or shorter duration, so account for this
                    # store mfcc for segment if it has the expected length
                    if len(mfcc) == expected_num_mfcc_vectors_per_segment:
                        data["mfcc"].append(mfcc.tolist())
                        data["labels"].append(i-1) # becuase first iteration was for dataset path itself so i-1
                        print(f"{file_path}, segment: {s}")
                    
    with open(json_path, "w") as fp:
        json.dump(data, fp, indent=4)
                    
                    
```

```python
save_mfcc(dataset_path=DATASET_PATH, json_path=JSON_PATH,
          n_mfcc=13, n_fft = 2048,
              hop_length=512, num_segments=10)
```
